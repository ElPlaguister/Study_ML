```python
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
for dirname, _, filenames in os.walk(os.path.join('data', 'car')):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
df = pd.read_csv(os.path.join(dirname, 'audi.csv'))
df.shape
```

```python
df.info()
```

```python
df.describe()
```

```python
ax = sns.boxplot(y = df['engineSize'])
```

```python
df.query('engineSize == 0')
```

```python
df = df.loc[df.engineSize != 0]
df.describe()
```

```python
from sklearn import preprocessing
```

```python
le = preprocessing.LabelEncoder()
le.fit(df.model)
le.classes_
m = le.transform(df.model)
```

```python
le.fit(df.transmission)
le.classes_
l = le.transform(df.transmission)
```

```python
le.fit(df.fuelType)
le.classes_
f = le.transform(df.fuelType)
```

```python
df['c_model'] = m
df['c_transmission'] = l
df['c_fuel'] = f
df.head()
```

```python
sns.heatmap(df.corr(), annot = True, fmt = '.1g')
```

```python
df2 = pd.get_dummies(df, prefix = ['fuel'], columns= ['c_fuel'])
df2 = pd.get_dummies(df, prefix= ['model'], columns = ['c_model'])
df2.head()
```

```python
df2 = df2.drop(['model', 'transmission', 'fuelType'], axis = 1)
```

```python
train, validate, test = np.split(df2, [int(.6 * len(df)), int(.7 * len(df))])
X_train = train.drop('price', axis = 1)
y_train = train.price
X_validate = validate.drop('price', axis = 1)
y_validate = validate.price
X_test = test.drop('price', axis = 1)
y_test = test.price
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
result = regressor.predict(X_validate)
y_pred_val = pd.Series(result)
```

```python
from sklearn import metrics
print('Mean Absolute Error:', metrics.mean_absolute_error(y_validate, y_pred_val))
print('Mean Squared Error: ', metrics.mean_squared_error(y_validate, y_pred_val))
print('Root Mean Squared Error', np.sqrt(metrics.mean_squared_error(y_validate, y_pred_val)))
```

```python
result2 = regressor.predict(X_test)
y_pred = pd.Series(result2)
print('Mean Absolute Error:', metrics.mean_absolute_error(y_pred, y_test))
print('Mean Squared Error: ', metrics.mean_squared_error(y_pred, y_test))
print('Root Mean Squared Error', np.sqrt(metrics.mean_squared_error(y_pred, y_test)))
```
